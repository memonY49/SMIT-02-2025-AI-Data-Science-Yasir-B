```python
student_data = pd.read_csv('students_five_classes.csv')
```

```python
student_data['obtain_marks'] = student_data.iloc[:,2:].sum(axis = 1)
```

```python
def grade(marks):
    if marks>80:
        return 'A'
    elif marks>70:
        return 'B'
    elif marks>50:
        return 'C'
    else:
        return 'F'
```

```python
student_data['total_marks'] = 100
```

```python
student_data['percentage'] = (student_data.obtain_marks / student_data.total_marks)*100
```

```python
student_data["grades"] = student_data.percentage.apply(grade)
```

```python
x = student_data.grades.value_counts().index
y = student_data.grades.value_counts().values
```

```python
bar_plot = plt.bar(x,y)
plt.bar_label(bar_plot)
```

```python
student_data['class'] = student_data['roll_no'].str.split('/').str[1]
```

```python
student_data
```

```python
classgroups = student_data.groupby('class')
```

```python
for cl in student_data['class'].unique():
    count = classgroups.get_group(cl).grades.value_counts()
    x = count.index
    y = count.values
    plt.figure(figsize=(10,10))
    plt.bar(x,y, color = ["red","yellow","blue","green"], alpha = 0.5, width = 1)
    plt.title(f"Class: {cl}")
    plt.xlabel("Grades")
    plt.ylabel("Student counts")
    plt.show()
```
