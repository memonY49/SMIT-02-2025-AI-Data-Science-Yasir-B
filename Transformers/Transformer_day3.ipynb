## Prompt Control Techniques

Control output using:
- Clear instructions
- Role specification
- Output format constraints
- Context boundaries

Example:
"You are a customer support agent. Respond politely in 3 sentences."



---

1. Input Prompt
"You are a customer support agent. Respond politely:\nCustomer: My order arrived late."

- This is prompt engineering.

- You are instructing the model to:

- Take the role of a customer support agent

- Use a polite tone

- \nCustomer: helps structure the conversation.

- The model will continue the text as if replying to the customer.

2. max_length=70

- Maximum number of tokens (not words) in the output.

- Includes input + generated text.

- Keeps the response short and controlled.

3. temperature=0.7

- Controls randomness / creativity.

- Lower values (e.g., 0.2) → more deterministic, safer responses.

- Higher values (e.g., 1.0) → more creative, varied responses.

- 0.7 is a good balance for customer support replies.

4. top_k=50

- The model considers only the top 50 most probable tokens at each step.

- Prevents very unlikely or nonsense words.

- Improves coherence.

5. top_p=0.9 (Nucleus Sampling)

- Instead of a fixed number, it selects tokens whose cumulative probability ≤ 90%.

- Adapts dynamically to the model’s confidence.

- Often produces more natural text than top_k alone.

6. do_sample=True

- Enables sampling-based generation.

- Required for temperature, top_k, and top_p to have an effect.

- If False, the model uses greedy decoding (always picks the most likely token).

```python
prompt = "My order arrived late."
output = generator(
    prompt,
    max_length=80,
    num_return_sequences=1
)

print(output[0]["generated_text"])
```

```python
system = "You are a customer support agent. Respond politely:\nCustomer: "
output = generator(
    system+"My order arrived late.",
    max_length=70,
    temperature=0.7,
    top_k=50,
    top_p=0.9,
    do_sample=True
)

print(output[0]["generated_text"])
```

## Reducing Hallucination Risks

Industry techniques:
- Use smaller temperature (≤ 0.7)
- Limit max tokens
- Provide context explicitly
- Prefer extractive QA over free generation
- Use retrieval-based systems


## Extractive Question Answering

Extractive QA:
- Answers only from provided context
- No new information is generated
- Highly reliable

Used in:
- Knowledge bases
- Policy documents
- Customer support FAQs


```python
qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad",
    device=device
)
```

## FAQ-Based Question Answering System

The model extracts answers **only from the context**.


```python
context = """
You are a qa bot and your name is 'ali from smit'.
My company name is smit and we sells some AI products in us market.
Our return policy allows customers to return products within 30 days of purchase.
Refunds are processed within 5 business days after receiving the product.
"""

question = "what is the name of your company?"

answer = qa_pipeline(
    question=question,
    context=context
)

print("Answer:", answer["answer"])
print("Confidence:", answer["score"])
```

## Customer Support Response Generator

Combines:

## Product Description Generator
Using mistralai
- Used in:
  - E-commerce platforms
  - Marketing automation
  - Content creation


```python
generator = pipeline(
    "text-generation",
    model="mistralai/Mistral-7B-Instruct-v0.2",
    device=device
)

prompt = "Write a professional 3-sentence product description for wireless noise cancelling headphones."

print(generator(prompt, max_new_tokens=100)[0]["generated_text"])
```

```python
prompt = (
    "You are a professional product copywriter.\n"
    "Write a professional product description.\n"
    "Product: Wireless Noise Cancelling Headphones.\n"
    "Write exactly 3 sentences."
)

output = generator(
    prompt,
    max_new_tokens=80,
    temperature=1,
    top_p=0.9,
    do_sample=True,
    eos_token_id=generator.tokenizer.eos_token_id
)

print(output[0]["generated_text"])
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_id = "meta-llama/Llama-3.1-8B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_8bit=True,
    torch_dtype=torch.float16
)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",

    torch_dtype=torch.float16
)
```

```python
prompt = "Explain transformer in datascience in simple terms."

inputs = tokenizer(
    prompt,
    return_tensors="pt"
).to(model.device)

with torch.no_grad():
    output = model.generate(
        **inputs,
        max_new_tokens=200,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )

print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
messages = [
    {"role": "system", "content": "You need to answer only on datascince subject. if you see anything which is not from datascince say you can only reply on questions from datascince."},
    {"role": "user", "content": "how to build a chiken biryani?"}
]

prompt = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        temperature=0.7,
        top_p=0.9,
        do_sample=True
    )

response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response)
```
