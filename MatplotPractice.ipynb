```python
import matplotlib.pyplot as plt
```

```python
x_values = [0, 1, 2, 3, 4, 5 ]
y_values = [0, 1, 4, 9, 16,25]
y2_values = [2, 3, 6, 12, 18,27]
# plt.figure(figsize=(5,10))
plt.plot(x_values,y_values,label = 'a', c = 'red', marker = '*',linestyle = '-.')
plt.plot(x_values,y2_values, label = 'b')
plt.text(1.2,1,"SMIT Class")
plt.title("SMIT Batch 2 Data Visualization")
plt.xlabel("X label")
plt.ylabel("Y label")
# plt.savefig('SMITFig.png')
plt.ylim(-50,50)
plt.xlim(-10,10)
plt.axhline(0, c='y')
plt.axvline(0, c = 'g')
plt.legend()
plt.show()
```
