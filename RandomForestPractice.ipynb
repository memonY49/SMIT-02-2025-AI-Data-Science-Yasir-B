```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
```

```python
df =  pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
df.drop('smoking_history', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['gender'] = encoder.fit_transform(df.gender)
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=40)
```

```python
x_train
```

```python
y_train
```

```python
RF_classifier = RandomForestClassifier(n_estimators=300)
```

```python
RF_classifier.fit(x_train,y_train)
```

```python
test_pred = RF_classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
print('test acc:',accuracy_score(y_test,test_pred))
```

```python
std = pd.read_csv('Datasets/students_adaptability_level_online_education.csv')
```

```python
std = std.select_dtypes(include=object).apply(encoder.fit_transform)
```

```python
x = std.iloc[:, :-1]
y = std.iloc[:, -1]
```

```python
std_x_train, std_x_test, std_y_train, std_y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
RF_C = RandomForestClassifier()
```

```python
grid_param = {"n_estimators": [200,400,500,1000],
             "criterion": ['gini','entropy'],
             "min_samples_split": [3,4,5,6,7]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
RF_Grid = GridSearchCV(estimator = RF_C,
                      param_grid = grid_param,
                      cv = 5)
```

```python
RF_Grid.fit(std_x_train,std_y_train)
```

```python
RF_Grid.best_params_
```

```python
RF_Grid.best_score_
```

```python
std_test_pred = RF_Grid.predict(std_x_test)
```

```python
print(classification_report(std_y_test,std_test_pred))
```
