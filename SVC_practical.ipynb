```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sklearn.svm import SVC
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df,iris.target,test_size=0.2,random_state=42)
```

```python
svc = SVC(C=5)
```

```python
svc.fit(x_train, y_train)
```

```python
x_test_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,x_test_pred))
```

# Import dataset from sklearn.datasets (Breast cancer)
## apply SVM(SVC) on dataset to predict cancer
- **Get Accuracy**
- **Get Precesion**
- **Get Recall**
- **Get F1_score**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
x = data.data
y = data.target
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
pipeline = Pipeline([('SC',StandardScaler()),('PCA',PCA(n_components=2)),('LR',LogisticRegression()),('SVC',SVC(C=4))])
```

```python
pipeline.fit(x_train,y_train)
```

```python
x_test_pred = pipeline.predict(x_test)
```

```python
print(accuracy_score(y_test,x_test_pred))
```
