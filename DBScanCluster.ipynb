```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
x,y = make_moons(n_samples=300, noise=0.10)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled.shape
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import DBSCAN
```

```python
db_clustring = DBSCAN(min_samples=4, eps= 0.5)
```

```python
db_clustring.fit(x_scaled)
```

```python
plt.scatter(x[:,0],x[:,1], c = db_clustring.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x_scaled,db_clustring.labels_)
```
