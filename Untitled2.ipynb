```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering
```

```python
x,y = make_moons(n_samples=500, noise=0.05)
scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
```

```python
plt.scatter(x_scale[:,0],x_scale[:,1])
```

```python
Kmeans = KMeans(n_clusters=2)
Agg_cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
Dbscan = DBSCAN(min_samples=4 ,eps=0.5)
```

```python
Kmeans.fit(x_scale)
Agg_cluster.fit(x_scale)
Dbscan.fit(x_scale)
```

```python
plt.scatter(x_scale[:,0],x_scale[:,1],c=Kmeans.labels_)
```

```python
plt.scatter(x_scale[:,0],x_scale[:,1], c = Agg_cluster.labels_)
```

```python
plt.scatter(x_scale[:,0],x_scale[:,1], c= Dbscan.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
print("Silhouette Score Kmeans:",silhouette_score(x_scale, Kmeans.labels_))
```

```python
print("Silhouette Score Agg:",silhouette_score(x_scale, Agg_cluster.labels_))
```

```python
print("Silhouette Score Dbscan:",silhouette_score(x_scale, Dbscan.labels_))
```
