```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.lineplot(iris,x = iris.sepal_length,y = iris.sepal_width,hue='species')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.lineplot(titanic,x = 'pclass', y = 'survived', hue = 'sex')
plt.scatter(titanic.pclass,titanic.survived)
```

```python
flight = sns.load_dataset('flights')
```

```python
sns.lineplot(flight, x= 'month', y = 'passengers', hue = 'year', markers='o',)
```

```python
sns.countplot(titanic,x = 'survived', hue = 'sex', )
```

```python
sns.barplot(titanic,x = 'embark_town', y ='fare', hue = 'alone')
plt.savefig("titanic_town_fare_alone.png")
```
