```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
trails = 100000
rolls = np.random.randint(1,7, size=trails)
value, counts = np.unique(rolls,return_counts= True)
p_count =  counts/trails
data = dict(zip(value,p_count))
plt.bar(value, p_count)
```
