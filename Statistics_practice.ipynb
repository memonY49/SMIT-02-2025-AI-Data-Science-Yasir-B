# Z - Score
- z_score = (xi-mean)/std

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = [1,4,5,7,3,9,100,5,8,105,5,8,8] 
q1, q3 = np.percentile(data, [25,75])
iqr = q3-q1
variance = np.var(data)
sd = np.std(data)
mean = np.mean(data)
print(f"Variance: {variance}")
print(f"STD: {sd}")
print(f"Mean: {mean}")
print(f"IQR: {iqr}")
print(q1,q3)

print(f"Lower fence: {q1-1.5*iqr}")
print(f"upper fence: {q3+1.5*iqr}")
print(sorted(data))
sns.boxplot(data)
```

# Type of Events
## 1. Simple Event:
   - Also known as an elementary event, a simple event is an event that consists of exactly one outcome.
   - ***For example***, when rolling a fair six-sided die, getting a 3 is a simple event.
## 2. Compound Event:
   - A compound event consists of two or more simple events.
   - ***For example***, when rolling a die, the event "rolling an odd number" is a compound event <br> 
    because it consists of three simple events: rolling a 1, rolling a 3, or rolling a 5.
## 3. Independent Events:
   - Two events are independent if the occurrence of one event does not <br> affect the probability of the occurrence of the other event.
   - ***For example***, if you flip a coin and roll a die, <br>the outcome of the coin flip does not affect the
    outcome of the die roll.
## 4. Dependent Events:
   - Events are dependent if the occurrence of one event affect the <br>probability of the occurrence of the other event.
   - ***For example***, if you draw two cards from a deck without replacement, the outcome of the <br> first draw affects the outcome of the second draw because there are fewer cards left in the
deck.
## 5. Mutually Exclusive Events: 
    - Two events are mutually exclusive (or disjoint) <br>if they cannot both occur at the same time.<br>
    ***For example***, when rolling a die, the events "roll a 2" and "roll a 4" are mutually exclusive <br>because a single roll of the die cannot result in both a 2 and a 4.
## 6. Exhaustive Events:
   - A set of events is exhaustive if at least one of the events must occur <br>when the experiment is performed.
   - ***For example***, when rolling a die, the events "roll an even number" and "roll an odd <br>number" are exhaustive because one or the other must occur on any roll.

# What is Probability
- In simplest terms, probability is a measure of the likelihood that a particular event will occur. It <br>is a fundamental concept in statistics and is used to make predictions and informed decisions <br>in a wide range of disciplines, including science, engineering, medicine, economics, and social
sciences.

**Probability is usually expressed as a number between 0 and 1, inclusive:**
- A probability of 0 means that an event will not happen.
- A probability of 1 means that an event will certainly happen.
- A probability of 0.5 means that an event will happen half the time (or that it is as likely to happen as not to happen).

# Empirical Probability Vs Theoretical Probability
## Empirical Probability
Empirical probability, also known as experimental probability, is a probability measure that is based on observed data, rather than theoretical assumptions. It's calculated as the ratio of the number of times a particular event occurs to the total number of trials. <br>
- No of times the event occurs/total no of trails<br>
***For example***,
  1. Suppose that, in our 100 tosses, we get heads 55 times and tails 45 times. What is the empirical
probability of getting a head.
  2. Let's say you have a bag with 50 marbles. Out of these 50 marbles, 20 are red, 15 are blue, and 15 are
green. You start to draw marbles one at a time, replacing the marble back into the bag after each draw. After 200 draws, you find that you've drawn a  red marble 80 times, a blue marble 70 times, and a green marble 50 times. What is the empirical probability of getting a red marble?
## Theoretical Probability
Theoretical (or classical) probability is used when each outcome in a sample space is equally likely to occur. If we denote an event of interest as Event A, we calculate the theoretical probability of that event as:
- Theoretical Probability of Event A = Number of Favourable Outcomes (that is, outcomes in Event A) / Total Number of Outcomes in the Sample Space
- Total no of events/total no of pssible outcomes<br>
***For Example***,
  1. Consider a scenario of tossing a fair coin 3 times. Find the probability of getting exactly 2 heads.
  2. Consider a scenario of rolling 2 dice. What is the probability of getting a sum = 7

# Random Variable
In the context of probability theory, a random variable is a function that maps the
outcomes of a random process (known as the sample space) to a set of real numbers.<br>
**Input:** The input to the function is an outcome from the sample space of a random
process.<br>
**Output:** The output of the function is a real number that we assign to each possible
outcome.
- The transformation from input to output in the function of a random variable is
determined by how we choose to define the random variable.
And the choice of how to define a random variable often depends on the specific aspects
of the random process (or event) that you're interested in studying.

# 2) Probability Density Function
A Probability Density Function (PDF) is a function that describes the relative likelihood of a continuous random variable taking on a particular value.
- Key Points:<br>
    1) For Continuous Variables
       - Unlike discrete variables (where we use a Probability Mass Function PMF), continuous variables dont have probability at exact points.
       - Instead, probability is given over intervals.
    2) Area Under Curve = Probability
       - The total area under the PDF curve = 1 (100%).
       - Probability that the variable falls in an interval [a, b] is the area under the curve between a and b.
# 3) Probability Mass Function
A Probability Mass Function (PMF) applies to discrete random variables.<br>
It gives the probability that the random variable is exactly equal to a specific value.
- Key Points:
    1) For Discrete Variables
       - Works when outcomes are countable (e.g., dice rolls, coin flips, number of customers).
    2) Definition
       - For a discrete random variable X, the PMF is:  (=)=()
       - where: ()0  and f(x)=1
    3) Examples
       - A fair die: P(X=x)=1/6,  x=1,2,3,4,5,6
       - A coin flip:P(X=Heads)=0.5,  P(X=Tails)=0.5
# 4) Comulative Density Function
The Cumulative Distribution Function (CDF) gives the probability that a random variable X is less than or equal to a certain value.
