```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
k = 3
```

```python
kmeans = KMeans(n_clusters=k)
```

```python
specie_label = kmeans.fit_predict(X)
```

```python
plt.scatter(X.iloc[:,0],X.iloc[:,1], c = specie_label)
```

```python
import seaborn as sns
```

```python
pg_data = sns.load_dataset('penguins')
```

```python
pg_data.dropna(axis=0, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Encoder =  LabelEncoder()
pg_data[["species","island","sex"]] = pg_data.select_dtypes(include=object).apply(Encoder.fit_transform)
```

```python
pg_data
```

```python
k_values = range(1,11)
```

```python
x = pg_data.iloc[:, 1:]
```

```python
intertia = []
```

```python
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    intertia.append(kmeans.inertia_)
```

```python
plt.plot(k_values,intertia, marker = 'o')
```
